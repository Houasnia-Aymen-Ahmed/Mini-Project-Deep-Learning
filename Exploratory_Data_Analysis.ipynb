# Import necessary libraries

```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.preprocessing import image
```

# Visualize random images from the dataset

```python
dataset_path = "Dataset/images/"
labels_df = pd.read_csv("Dataset/labels.csv")

plt.figure(figsize=(12, 8))

for i in range(6):
    # Choose a random image from the dataset
    img_file = np.random.choice(os.listdir(dataset_path))
    img_path = os.path.join(dataset_path, img_file)
    
    # Load and display the image
    img = image.load_img(img_path, target_size=(224, 224))
    
    # Assuming the label is present in the file name (e.g., "001.jpg")
    image_label = img_file.split('.')[0]
    
    # Get the corresponding label from the CSV file
    label_info = labels_df[labels_df['image'] == f"{image_label}.jpg"]
    if not label_info.empty:
        actual_label = label_info.iloc[0]['label']
        plt.subplot(2, 3, i+1)
        plt.imshow(img)
        plt.title(f"Label: {actual_label} - {image_label}")

plt.show()
```
